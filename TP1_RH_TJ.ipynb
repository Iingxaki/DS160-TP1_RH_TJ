# Explore Data Analysis (EDA Process)

### Identify the which year most of the LoL games posted in the data

```python
data['Year'].value_counts()
```

```python
data['Year'].value_counts().sort_values(ascending=False)\
.plot(kind='bar',figsize=(8,6),title="Year wise LoL games in the data")
plt.xlabel("Year")
plt.ylabel("Count of values")
plt.show()
```

### Observations:
#### bar chart
- Year 2017 has most of the matches posted compare to the other years
- In 2014 we received the least amount of matches compared to other years


### Identify the which tournament most of the LoL games posted in the data

```python
data['League'].value_counts()
```

```python
data['League'].value_counts().sort_values(ascending=False)\
.plot(kind='bar',figsize=(8,6),title="Region's League wise LoL games in the data")
plt.xlabel("Region's League")
plt.ylabel("Count of values")
plt.show()
```

### Observations:
#### bar chart
- LCK has the highest amount of matches posted.
- RR has the lowest amount of matches posted.


### Identify the which type of matches have in the data set using the pie chart

```python
data['Type'].value_counts()
```

```python
data['Type'].value_counts().sort_values(ascending=False)\
.plot(kind='pie',figsize=(8,6),title="To Visualize the match types in the data",explode=[0.05,0.04,0.05,0.06,0.07],
    labels=['Season','Playoffs','International','Promotional','Regional'],
    autopct='%1.2f%%')
plt.show()
```

### Observations:

#### Pie Chart
- Most of the matches posted were seasonal.
- Least matches posted were regional.

### Create histogram to understanding game length with region

```python
sns.histplot(data=data,x='gamelength' ,hue='League',kde=True)
plt.title("Understanding game length by Region")
plt.xlabel("game length")
plt.ylabel("Frequency")
plt.show()
```

### Observations:
- From the histogram we observed some interesting things about game length along with region.
- In the histogram most of the game length went from 25-45 min. It seems like LCK also had more games posted in that same range.

### Create histogram to understanding game length with match type

```python
sns.histplot(data=data,x='gamelength' ,hue='Type',kde=True)
plt.title("Understanding game length by match type")
plt.xlabel("game length")
plt.ylabel("Frequency")
plt.show()
```

### Observations:
- From the histogram we observed some interesting things about game length along with match type.
- In the histogram most of the game length went from 25-45 min. It seems like Seasonal matches also had more games posted in that same range.

### Visualize the bar chart the average game length for blue team's result win or loss

```python
data.groupby(['bResult'])['gamelength'].mean().sort_values(ascending=False)\
.plot(kind='bar',title="Sum of the game length in with blue team result",figsize=(8,6),color=['#F7F2E0','#0B3B0B'])
plt.xlabel("blue team result")
plt.ylabel("mean of game length")
plt.show()
```

### Visualize the bar chart the average game length for blue team's Jungle Champion picked

```python
data.groupby(['blueJungleChamp'])['gamelength'].mean().sort_values(ascending=False)\
.plot(kind='bar',title="Sum of the game length in with blue Jungle Champ",figsize=(8,6),color=['#4B610B','#610B5E','#E3CEF6','#F5F6CE','#D0A9F5','#F6E3CE'])
plt.xlabel("blue team Jungle Champion")
plt.ylabel("mean of game length")
plt.show()
```

### Observations:
- From the chart we observed that Xin zhao has the highest average game length.
- From the chart we observed that Leblanc has the lowest average game length.
